```python
import pandas as pd
import json
```

### 1. jsonファイル読み込み

```python
json_open = open('pokemon_skills_utf8.json', 'r')
pk_json = json.load(json_open)
pk_json
```

### 2. すべての技一覧を確認(dictionary作成)

```python
remember_skills = []
for i in range(1, len(pk_json)+1):
    pk = pk_json[str(i)]
    for skill in pk['skills']:
        remember_skills.append(skill)

skill_dictionary = set(remember_skills)

len(skill_dictionary)
```

```python
skill_dictionary
```

### 3. 類似度計算用のDataFrame作成

#### 3.1 (ポケモン, 覚える技)のdf

```python
df = pd.DataFrame()

for i in range(1, len(pk_json)+1):
    pk = pk_json[str(i)]

    df_record = pd.DataFrame(
        {
         'id': i,
         'name': pk['name'],
         'skill': pk['skills']                
        }
    )
    df = pd.concat([df, df_record])
        
df.head()
```

#### 3.2 覚える技をOne-Hotに変更

```python
df_skill_dic = pd.get_dummies(df['skill'])

df2 = pd.concat([df[['id', 'name']],df_skill_dic], axis=1)
df2.head()
```

```python
df3 = df2.groupby(['id', 'name']).sum().reset_index(drop=False)
df3 = df3.set_index('id')
df3.head()
```

### 4. 類似度を算出

覚える技の空間座標上の距離で類似度を数値化

ポケモンをIDで指定

id ==1 (フシギダネの場合)

```python
pk_id = 1
df3_tmp = df3.drop(['name'], axis=1).astype(float)
df_distance = df3_tmp - df3_tmp.loc[pk_id]
df_distance = df_distance.abs()
df_distance.sum(axis=1).sort_values()
```
